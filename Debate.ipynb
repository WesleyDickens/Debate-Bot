```python
from openai import OpenAI
import time
import os
from dotenv import load_dotenv
```

```python
load_dotenv()

api_key_1 = os.getenv('OPENAI_API_KEY_1')
api_key_2 = os.getenv('OPENAI_API_KEY_2')

client1 = OpenAI(api_key = api_key_1)
client2 = OpenAI(api_key = api_key_2)
```

```python
def conversation(input_text, original_context):
    if input_text != original_context:
        message = original_context + " " + input_text
    else:
        message = input_text

    bot_1_response = client1.chat.completions.create(
        model="gpt-4",
        messages=[
            {"role": "system", "content":"You are an atheist trying to convince a user to believe in your worldview. Responses should be 2 sentences max. Your reply should directly reference the person's response."},
            {"role": "assistant", "content": message}
        ],
        presence_penalty=0.5,

    )

    print(f"Atheist Bot: {bot_1_response.choices[0].message.content}\n")

    message = original_context + bot_1_response.choices[0].message.content

    bot_2_response = client2.chat.completions.create(
        model="gpt-4",
        messages=[
            {"role": "system", "content":"You are a Christian trying to convince a user to believe in your worldview. Responses should be 2 sentences max. Your reply should directly reference the person's response."},            
            {"role": "user", "content": message}
        ],
        presence_penalty=0.5,

    )

    print(f"Christian Bot: {bot_2_response.choices[0].message.content}\n")

    return bot_2_response.choices[0].message.content


def main():
    # Start the conversation with a question
    input_text = "Is there a God?"
    original_input = input_text
    # Limit conversation to 20 responses from each bot
    for i in range(20):
        input_text = conversation(input_text=input_text, original_context=original_input)
        time.sleep(3)


if __name__ == '__main__':
    main()
```
